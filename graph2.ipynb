```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import size
import seaborn as sns
```

Probability of 1 = 0.7

```python
df = pd.read_csv("data/01-thresholdRecursive-Barvinok70.csv")
df = df.loc[df["Error"]<2]

fig, ax = plt.subplots(figsize=(18,8))
sns.set(style="whitegrid")
sns.violinplot(x=df["N"], y=abs(df["Error"]))
ax.set_ylim([-1,7])

plt.title("Relative Error in the Permanent Calculated by Recursive Barvinok on Matrices with Probability Density = 0.7")
plt.savefig("recursiveBarvinok7.png")
```

```python
df = pd.read_csv("data/recursiveData/01-thresholdRecursive-Barvinok70.csv")
df2 = pd.read_csv("data/recursiveData/01-thresholdRecursive-Barvinok80.csv")

x = [10, 11, 12, 13, 14, 15]
y2 = []
y3 = []


for i in x:
    y2.append(df.loc[(df["N"]==i), "Epochs"].mean())
    y3.append(df2.loc[(df2["N"]==i), "Epochs"].mean())

fig, ax = plt.subplots(figsize=(12,6))

plt.plot(x, y2, label="p = 0.7", marker='o')
plt.plot(x, y3, label="p = 0.8", marker='o')
ax.set_xlabel("Size of Matrix N")
ax.set_ylabel("Number of Epochs for Convergence")
plt.title("Average Number of Epochs for Convergence by the algorithm")
ax.legend()
plt.savefig("images/recursiveBarvinok/epochs")
```
