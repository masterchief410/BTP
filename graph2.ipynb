```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import size
import seaborn as sns
```

Probability of 1 = 0.7

```python
df = pd.read_csv("data/01-thresholdRecursive-Barvinok70.csv")

fig, ax = plt.subplots(figsize=(18,8))
sns.set(style="whitegrid")
sns.violinplot(x=df["N"], y=abs(df["Error"]))
ax.set_ylim([-1,7])

plt.title("Relative Error in the Permanent Calculated by Recursive Barvinok on Matrices with Probability Density = 0.7")
plt.savefig("recursiveBarvinok7.png")
```

```python
df = pd.read_csv("data/new01-RAP70.0.csv")

fig, ax = plt.subplots(figsize=(18,8))
sns.set(style="whitegrid")
sns.violinplot(x=df["N"], y=abs(df["Error"]))
ax.set_ylim([-1,7])
plt.title("Relative Error in the Permanent Calculated by RAP Method on Matrices with Probability Density = 0.7")
plt.savefig("rap7.png")
# sns.swarmplot(x=df["N"], y=df["Error"])
```
