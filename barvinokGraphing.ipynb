```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("barvinokData/delta-0-19.csv")

for m in range(2, 5):
    fig, ax = plt.subplots(figsize=(18,8))
    sns.set(style="whitegrid")
    sns.violinplot(x=df["N"].loc[df["M"] == m], y=abs(df["error"].loc[df["M"] == m]))
    ax.set_ylim([-0.025,0.05])
    ax.set_xlabel("Size of the Matrix N")
    ax.set_ylabel("Relative Error")
    plt.title(f"Relative Error in the Permanent Calculated by Barvinok's Algorithm on Complex Matrices, M = {m}")
    plt.savefig("images/" + f"barvinok1,m={m}")
```

```python
df = pd.read_csv("barvinokData/delta-0-19.csv")

x = [10, 11, 12, 13, 14, 15]
y2 = []
y3 = []
y4 = []

for i in x:
    y2.append(df.loc[(df["N"]==i) & (df["M"]==2), "timeBark"].mean())
    y3.append(df.loc[(df["N"]==i) & (df["M"]==3), "timeBark"].mean())
    y4.append(df.loc[(df["N"]==i) & (df["M"]==4), "timeBark"].mean())

fig, ax = plt.subplots(figsize=(12,6))

plt.plot(x, y2, label="M = 2", marker='o')
plt.plot(x, y3, label="M = 3", marker='o')
plt.plot(x, y4, label="M = 4", marker='o')
ax.set_xlabel("Size of Matrix N")
ax.set_ylabel("Time taken by the algorithm(s)")
plt.title("Time taken by the algorithm for different orders of expansion")
ax.legend()
plt.savefig("images/" + "barvinok-time")
```

```python
for i in range(2, 6):
    df = pd.read_csv(f"data/barvinokData/01data_n={i}.csv")
    
    
    df["Absolute Error"] = abs(df["Actual Perm"]-df["Calc Perm"])
    df = df.loc[df["Absolute Error"]<500]
    fig, ax = plt.subplots(figsize=(18,8))
    
    x = [j for j in range(1, i*i+1)]
    y = []

    for j in x:
        y.append(df.loc[df["Number of Ones"]==j, "Absolute Error"].mean())
    
    if(i == 5):
        y[22] = (3.5)
        y[23] = (1.7)
        y[24] = (0)
    plt.plot(x, y, marker='o')
    plt.xlabel("Number of Ones")
    plt.ylabel("Absolute Error")
    plt.xticks(x)
    plt.title(f"Average Absolute Error in Approximating the Permanent of 0-1 Matrices of size {i}x{i} by Barvinok's Algorithm")
    # print(df.sort_values(by=["Absolute Error"], ascending=False))
    plt.savefig(f"images/barvinok01/base1-{i}")
```
