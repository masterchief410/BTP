```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")

x9 = df["del"][0:9]
y9 = df["Relative Error"][0:9]

x10 = df["del"][9:]
y10 = df["Relative Error"][9:]
plt.plot(x9, y9, label = "N = 9, M = 3", marker='o')
plt.plot(x10, y10, label = "N = 10, M = 3", marker='s')
plt.legend()
plt.xlabel("Delta")
plt.ylabel("Relative Error")
```

```python
df2 = pd.read_csv("data.csv")

x = df2["N"][-5:]
y1 = df2["Time Approximation"][-5:]
y2 = df2["Time run calculation"][-5:]

plt.plot(x, y1, label = "Approximation Algorithm at M = 3", marker='o', color='g')
plt.plot(x, y2, label = "Permanent Calculation", marker='s', color='r')
plt.legend()
plt.xlabel("Size of Matrix (N)")
plt.ylabel("Time (s)")
```

```python
df2 = pd.read_csv("data.csv")

x = df2["N"][18:26]
y1 = df2["Relative Error"][18:26]
y2 = df2["Relative Error"][26:34]
y3 = df2["Relative Error"][34:42]

plt.plot(x, y1, label = "M = 3 and delta = 0.1", marker='o', color='y')
plt.plot(x, y2, label = "M = 3 and delta = 0.5", marker='s', color='r')
plt.plot(x, y3, label = "M = 3 and delta = 0.9", marker='v', color='g')
plt.legend()
plt.xlabel("Size of Matrix (N)")
plt.ylabel("Relative Error")
```
