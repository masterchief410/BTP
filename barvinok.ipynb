```python
import numpy as np
import itertools
import math
```

```python
# Parameters that can be varied N, M, delta

# 3 to 13
N = 6

# 1 to min(5, N)
M = 2

# delta 0.1 to 0.9 (0.1 intervals)
# delta = 0.9

# N x N matrix with all ones
J = np.ones((N, N))

# N x N matrix with random numbers from -delta to delta
A = np.round(np.random.rand(N, N))
A = A*2
A
```

```python
# g(z) = per (J + z(A-J))
# f(z) = ln (g(z))

# g[k] = delk/del zk g(z) at z = 0
# f[k] = delk/del zk f(z) at z = 0

# f(1) = f(0) + sigma(1/k! * f[k])
```

```python
import time
startTime = time.time()

f = np.zeros(M+1)
g = np.zeros(M+1)

tempPerm = np.array(range(1, N+1))
g[0] = math.factorial(N)

for k in range(1, M+1):
    all = list(itertools.combinations(tempPerm, k))
    allCombinations = [all,all]
    allProd = list(itertools.product(*allCombinations))

    for each in allProd:
        prod = 1.
        for i in range(0, k):
            prod = prod * (A[each[0][i]-1][each[1][i]-1]-1)
        print(prod, k)
        g[k] = g[k] + prod
    g[k] = g[k] * math.factorial(N-k)
    
# g[k] = sigma(j=0:k-1) (k-1Cj)*(g[j])*(f[k-j])
# f[k] = (g[k] - sigma(j=1:k-1) k-1Cj * g[j] * f[k-j])/g[0]*k-1

f[0] = math.log(g[0])

for k in range(1, M+1):
    sigma = 0
    for j in range(1, k):
        sigma = sigma + math.comb(k-1, j)*g[j]*f[k-j]
    f[k] = (g[k] - sigma)/(g[0]*k)

ans = f[0]

for i in range(1, M+1):
    ans = ans + (f[i]/math.factorial(i))

calculatedPerm = math.exp(ans)
calculatedPerm

algoTime = time.time()-startTime
```

```python
startTime = time.time()
permanent = 0

tempPerm = np.array(range(0, N))
for each in list(itertools.permutations(tempPerm)):
    product = 1
    for i in range(0, N):
        product = product * A[i][each[i]]
    permanent = permanent + product
ogTime = time.time()-startTime
```

```python
relative_error = abs(permanent - calculatedPerm)/permanent
print(algoTime, ogTime, calculatedPerm, permanent, relative_error)

import pandas as pd
df = pd.DataFrame(columns=["N","M","delta","Time for Approximation","Time to run actual calculation", "Approximated Permanent", "Calculated Permanent", "Relative Error"])

df["N"] = [N]
df["M"] = [M]
df["delta"] = [1]
df["Time for Approximation"] = [algoTime]
df["Time to run actual calculation"] = [ogTime]
df["Approximated Permanent"] = [calculatedPerm]
df["Calculated Permanent"] = [permanent]
df["Relative Error"] = [relative_error]

df.to_csv("data/data.csv", mode='a', index=False, header=False)
```
